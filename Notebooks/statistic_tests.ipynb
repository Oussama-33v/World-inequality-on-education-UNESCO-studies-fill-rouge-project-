this statistical test called chi2 countingecy is to see or relise if there is any depandency between categorial variables

```python
alpha = 0.05
df_categ  = df[['sex', 'location', 'wealth']]
for cat in df_categ:
    countingcy = pd.crosstab(df['region_group'], df_categ[cat])
    chi2_status, p, dof, ex = chi2_contingency(countingcy)
    if p < alpha:
        print("this variables are depandent")
    else:
        print("this varibles are indepandent")
print("finish")
print(p)
```

```python
alpha = 0.05
df_categ  = df[['sex', 'location', 'wealth']]
for cat in df_categ:
    countingcy = pd.crosstab(df['income_group'], df_categ[cat])
    chi2_status, p, dof, ex = chi2_contingency(countingcy)
    if p < alpha:
        print("this variables are depandent")
    else:
        print("this varibles are indepandent")
print("finish")
```

this tow tests called ANOVA one way it  used to comper the means of three or moor nemeric groups

```python
f_stats, p_value = stats.f_oneway(
    df['comp_prim_v2_m'],
    df['comp_lowsec_v2_m'], 
    df['comp_upsec_v2_m']
)

alpha = 0.05
if p_value < alpha:
    print("Significant difference exists; at least one group mean is different.")
else:
    print("No significant difference; groups are statistically similar.")
```

```python
f_stats, p_value = stats.f_oneway(
    df['eduout_prim_m'],
    df['eduout_lowsec_m'], 
    df['eduout_upsec_m']
    )

alpha = 0.05
if p_value < alpha:
    print("Significant difference exists; at least one group mean is different.")
else:
    print("No significant difference; groups are statistically similar.")
```
