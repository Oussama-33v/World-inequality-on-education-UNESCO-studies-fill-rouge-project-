```python
# bar plot presant the education compliton rate per region groups
df_long = df.melt(
    id_vars=['region_group'], 
    value_vars=['comp_prim_v2_m', 'comp_lowsec_v2_m', 'comp_upsec_v2_m'], 
    var_name='Education Level', 
    value_name='Completion Rate'
)
plt.figure(figsize=(11,5))
sns.barplot(data = df_long, x = 'region_group', y = 'Completion Rate', hue = 'Education Level', errorbar= None)
plt.xticks(rotation=45, ha='right')
plt.title("education completion rate per region groups")
plt.xlabel("Region Groups")
```

```python
# this graph presant the Percentage of children of primary school age who are not in school by region
df_long = df.melt(
    id_vars=['region_group'],
    value_vars=['eduout_prim_m', 'eduout_lowsec_m', 'eduout_upsec_m'], 
    var_name='Education out Level', 
    value_name='eduout Rate'
)
plt.figure(figsize=(11,5))
sns.barplot(data = df_long, x = 'region_group' , y = 'eduout Rate', hue = 'Education out Level', errorbar= None)
plt.xticks(rotation=45, ha='right')
plt.title("education out level rate by region")
```
